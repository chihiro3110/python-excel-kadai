```python
import pandas as pd


def main():
    # ===== 元データ =====
    data = [
        {"日付": "2023-05-17", "社員名": "山田", "売上": 100, "部門": "メーカー"},
        {"日付": "2023-05-18", "社員名": "佐藤", "売上": 200, "部門": "代理店"},
        {"日付": "2023-05-19", "社員名": "鈴木", "売上": 150, "部門": "メーカー"},
        {"日付": "2023-05-20", "社員名": "田中", "売上": 300, "部門": "商社"},
        {"日付": "2023-05-21", "社員名": "高橋", "売上": 250, "部門": "代理店"},
    ]

    # ===== DataFrame作成 =====
    df = pd.DataFrame(data)

    # ===== 全体の平均売上を算出 =====
    overall_avg = df["売上"].mean()

    # 各行に平均売上を追加
    df["平均売上"] = overall_avg

    # ===== 業績ランク判定 =====
    def rank(sales):
        if sales >= overall_avg + 50:
            return "A"
        elif sales >= overall_avg:
            return "B"
        else:
            return "C"

    df["業績ランク"] = df["売上"].apply(rank)

    # ===== 列順を指定 =====
    df = df[["日付", "社員名", "売上", "部門", "平均売上", "業績ランク"]]

    # ===== Excel出力 =====
    df.to_excel("業績.xlsx", index=False)

    print("業績.xlsx を作成しました")
    print(f"全体の平均売上: {overall_avg}")


if __name__ == "__main__":
    main()
```
